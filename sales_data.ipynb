```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_data.csv",encoding='unicode_escape')
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.ndim
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.head()
```

```python
dict={0:"unmarried",1:"married"}
df["Marital_Status"]=df["Marital_Status"].map(dict)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
plt.figure(facecolor="red")
x=df.groupby(['Gender'])['Amount'].mean().reset_index()
ax=sns.barplot(x="Gender",y="Amount",data=x)
plt.xlabel("Gender")
plt.ylabel("Amount")
for bar in ax.containers:
    ax.bar_label(bar, label_type='edge')

plt.show()
print("not such effect found, both males and females are doing around same amount of shopping")
```
