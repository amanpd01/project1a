```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_excel("student_data.xls")
```

```
df.head()
```

```
df.info()
```

```
df.isnull()
```

```
df.describe()
```

```
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```
df.head()
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="Gender",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("Gender")
plt.show()
print("males are scoring good marks in math subject and females are scoring good marks in the other subjects")

```

```
plt.figure(facecolor="grey")
sns.heatmap(pd.pivot_table(data=df,index="EthnicGroup",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("EthnicGroup")
plt.show()
print("group d and group e students are scroring good marks in all the subjects")
```

```
plt.figure(facecolor="yellow")
sns.heatmap(pd.pivot_table(data=df,index="ParentEduc",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("ParentEduc")
plt.show()
print("high and some high schools students are getting low marks compared to the other students who have good degree")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="LunchType",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("LunchType")
plt.show()
print("students with standard lunch type are scoring good marks")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="TestPrep",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("TestPrep")
plt.show()
print("students who completed their test preparation are scoring good marks")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="ParentMaritalStatus",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("ParentMaritalStatus")
plt.show()
print("marital status has no such effects in scoring high marks in any of the subjects")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="PracticeSport",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("PracticeSport")
plt.show()
print("students who practice sports regularly or even sometimes are getting slightly higher marks in all subjects")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="IsFirstChild",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("IsFirstChild")
plt.show()
print("is first child has no effect in students marks")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="NrSiblings",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("NrSiblings")
plt.show()
print("having less or more siblings making no effect in students marks")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="TransportMeans",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("TransportMeans")
plt.show()
print("students who are going to schools with schools bus or private bus are scoring around same marks in their exams")
```

```
plt.figure(facecolor="green")
sns.heatmap(pd.pivot_table(data=df,index="WklyStudyHours",values=["MathScore","ReadingScore","WritingScore"]),annot=True,linewidth=0.25)
plt.ylabel("WklyStudyHours")
plt.show()
print("weekly study hours has no effect in students marks")
```
