```python
#19
plt.figure(figsize=(10,3),facecolor="cyan")
ax=sns.barplot(x="PURPOSE*",y="MILES*",data=df)
plt.xlabel("purpose")
plt.ylabel("miles")
plt.title("purpose vs miles")
for i in ax.containers:
  ax.bar_label(i)
plt.show()
```

```python
#20
df["MILES*"].groupby(df["PURPOSE*"]).sum()
```

```python
#21
plt.figure(figsize=(10,3),facecolor="magenta")
ax=sns.countplot(x="CATEGORY*",data=df)
plt.xlabel("category")
for i in ax.containers:
  ax.bar_label(i)
plt.show()
```
